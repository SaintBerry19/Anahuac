```python
import random
import string

fig = plt.figure()
N, bins, patches = plt.hist(df['Under $1.9'],alpha = 0.4, label = "Country Population",color = "lightblue", ec="blue", lw=2)
for i in range(len(N)):
   patches[i].set_facecolor("#" + ''.join(random.choices("ABCDEF" + string.digits, k=6)))
plt.title("Frecuencia de Países vs % Población con Ingreso Menor a 1.9 USD/día")
plt.xlabel("% Población con Ingreso Menor a 1.9 USD/día")
plt.ylabel("Frecuencia de Países")
plt.savefig("Hist_190_RobertoMoraBalderas.png")
plt.show()
```

```python
indicador=df.columns
indicador=indicador[1:4]
contador=0
for i in indicador:
    contador=contador+1
    fig = plt.figure(figsize=(10, 9))
    sns.boxplot(y = i,  x = "Continent", data = df)
    plt.grid(True)
    title = str(i) + " By Continent"
    plt.title(title)
    plt.ylabel(f"% Population {i}/day")
    plt.savefig(f'Cajas_190_RobertoMoraBalderas_{contador}.png')
    plt.show()
```
