```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
import numpy as np
import pandas as pd
```

## Obtención del conjunto de los datos.

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

#### Instrucción 2: 
<p style='text-align: justify;'>Reproduce la siguiente celda de código para construir un dataframe (<b>df</b>) con la información de nuestro conjunto de datos, posteriormente incluye dos celdas de tipo <b>code</b> donde utilices la instrucción <b>head()</b> para visualizar los primeros diez elementos del conjunto de datos y la instrucción <b>describe()</b> para obtener información estadística básica.</p>

```python
df = pd.DataFrame(wine.data, columns = wine.feature_names)
```

#### Instrucción 4: 
<p style='text-align: justify;'>En la siguiente celda escribe las instrucciones adecuadas para construir un histograma a partir de la columna titulada "<b>alcohol</b>". Utiliza el parámetro <b>bins = 20</b>, para construir tu gráfico.</p>

```python
plt.hist(df['alcohol'], bins=20)
plt.title('Histograma de Alcohol')
plt.xlabel('Contenido de Alcohol')
plt.ylabel('Frecuencia')
plt.show()
```

#### Instrucción 6: 
<p style='text-align: justify;'>Utiliza la instrucción <b>corr()</b> para obtener la correlación de variables de nuestro dataframe. Guarda esta correlación en una variable llamada <b>corr</b>. E imprime la correlación con ayuda de la instrucción <b>print()</b>.</p>

```python
corr=df.corr()
print(corr)
```

#### Instrucción 7: 
<p style='text-align: justify;'>Construye un mapa de calor con la información de la matriz de correlaciones obtenida en la instrucción anterior. <b>Nota:</b> puedes usar los parámetros que quieras en el mapa de calor siempre y cuando el resultado sea legible, pero se te recomiendan los siguiente figsize =(10,8), linewidth = 0.5, annot = True, cmap = "YlGnBu".</p>

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, linewidth=0.5, annot=True, cmap="YlGnBu")
plt.title('Mapa de Calor de la Matriz de Correlaciones')
plt.show()
```

#### Instrucción 9: 
<p style='text-align: justify;'>Utilizando la instrucción <b>sns.joinplot</b> construye una visualización de datos donde se evidencie la relación entre el grado de alcohol y la intensidad del color del vino. <b>Nota:</b> Quizás sea conveniente grabar en una variable <b>x</b> a la columna del grado de alcohol (x = df['alcohol']) y en otra variable <b>y</b> la columna con los datos de la intensidad del color (y = df['color_intensity']). </p>

```python
x = df['alcohol']
y = df['color_intensity']
sns.jointplot(x=x, y=y, kind="scatter")
plt.show()
```
