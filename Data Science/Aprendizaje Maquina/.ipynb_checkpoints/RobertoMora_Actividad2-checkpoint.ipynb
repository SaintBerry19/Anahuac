#### Instrucción 2: 
<p style='text-align: justify;'>Reproduce la siguiente celda para incluir en nuestro dataframe a la variable que buscaremos predecir con nuestro modelo de regresión.</p>

```python
df['target'] = diabetes.target
df.head(10)
```

#### Instrucción 3: 
<p style='text-align: justify;'>Utiliza la instrucción <b>corr()</b> para obtener la correlación de variables de nuestro dataframe. Guarda esta correlación en una variable llamada <b>corr</b>.</p>

<p style='text-align: justify;'>Construye un mapa de calor con la información de la matriz de correlaciones obtenida en la instrucción anterior. <b>Nota:</b> puedes usar los parametros que quieras en el mapa de calor siempre y cuando el resultado sea legible, pero se te recomiendan los siguiente figsize =(10,8), linewidth = 0.5, annot = True, cmap = "YlGnBu".</p> </p>

```python
corr = df.corr()
sns.set(style="ticks", color_codes=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True, linewidths=.5, cmap="YlGnBu")
```

#### Instrucción 4: 
<p style='text-align: justify;'>De forma similar a lo que utilizamos en nuestra libreta practica titulada "<b>Regresión Lineal y Regresión Múltiple</b>" obten un arreglo donde se muestren las caracteristicas más relevantes y grabalas en una variable llamada <b>relevant_features</b>. Para este ejercicio considerea como una caracteristica relevante a toda aquella que tenga una correlación mayor al 40%.</p>

```python
cor_target = abs(corr["target"])
relevant_features = cor_target[cor_target>0.5]
relevant_features
```

#### Instrucción 5: 
<p style='text-align: justify;'>En la siguiente sección de nuestra libreta  analizaremos a la variable "<b>Body mass index</b>" (bmi). Crea una variable <b>x</b> donde grabes la columna <b>bmi</b> y una variable <b>y</b> donde almacenemos la información de la columna <b>target</b>. Para comprobar que lo anterior sea un exito imprime cada una de las nuevas variables. <b>Nota:</b> es muy recomendable a cada una de las variables les des la forma de una columna de datos con ayuda de la instrucción <b>values.reshape(-1, 1)</b>.</p>

```python
x = df["bmi"].values.reshape(-1, 1)
y = df["target"].values.reshape(-1, 1)
```

```python
print(x)
```

```python
print(y)
```

#### Instrucción 6: 
<p style='text-align: justify;'>Repoduce la siguiente celda, para divir el conjunto de datos en entrenamiento y prueba. Una vez que la reproduzcas la celda completa la información solicitada en la celda markdown.</p>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=100)
```

```python
print("Tamaño del conjunto de datos: ", x.size)
print("Tamaño del conjunto de entrenamiento: ", x_train.size)
print("Tamaño del conjunto de prueba: ", x_test.size)
```

#### Instrucción 7: 
<p style='text-align: justify;'>Con ayuda de la paquetería de sklearn.linear_model construye un modelo de regresión lineal que se ajuste a nuestros datos de entrenamiento y utiliza el modelo para obtener los dos siguientes conjuntos de respuestas: "<b>y_train_predict</b>" y "<b>y_test_predict</b>".</p>

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x_train,y_train)

y_train_predict = reg.predict(x_train)
y_test_predict = reg.predict(x_test)
```

#### Instrucción 8: 
<p style='text-align: justify;'>Utilizando la función <b>plt.scatter</b> construye una grafica donde se muestren el conjuto de entrenamiento en color azul, el conjunto de prueba en naranja y una línea recta roja donde se muestre la predicción devuelta por nuestro modelo.</p>

<p style='text-align: justify;'><b>Nota 1:</b> apoyate en el ejemplo de la libreta titulada "<b>Regresión Lineal y Regresión Múltiple</b>" en caso de que tengas algún problema o duda con esta sección.</p>

<p style='text-align: justify;'><b>Nota 2:</b> Al momento de construir las x_plot para graficar la línea recta creada por el modelo es conveniente que uses valores de inicio y de fin muy cercanos al valor minimo y maximo de la variable en cuestion.</p> 
- Valor minimo de la variable bmi: -9.027530e-02
- Valor maximo de la variable bmin: 1.705552e-01

```python
plt.scatter(x_train, y_train, c = "blue", alpha = 0.2)
plt.scatter(x_test, y_test, c = "orange", alpha = 0.8)

x_plot = np.linspace(-9.027530e-02,1.705552e-01).reshape(-1, 1)
y_plot = reg.predict(x_plot)

plt.plot(x_plot, y_plot,"r--", lw = 4)
plt.show()
```

#### Instrucción 9: 
<p style='text-align: justify;'>Reproduce la siguiente celda para obtener la metrica r2_score tanto para el conjunto de entrenamiento como en el de prueba.</p>

```python
from sklearn.metrics import r2_score
print("Entrenamiento", r2_score(y_train, y_train_predict))
print("Prueba", r2_score(y_test, y_test_predict))
```

#### Instrucción 10: 
<p style='text-align: justify;'>Repite las instrucciones cinco hasta la nueve pero ahora sustituye la variable "<b>Body mass index</b>" (bmi) por la variable "<b>BAverage blood pressure</b>" (bp). Una vez que termines de realizar las instrucciones, agrega en una celda de tipo <b>markdown</b> que significa que la metrica r2_score sea menor para esta variable con respecto a la variable bmi.</p>

```python
x = df["bp"].values.reshape(-1, 1)
y = df["target"].values.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=100)
reg = LinearRegression().fit(x_train,y_train)

y_train_predict = reg.predict(x_train)
y_test_predict = reg.predict(x_test)
plt.scatter(x_train, y_train, c = "blue", alpha = 0.2)
plt.scatter(x_test, y_test, c = "orange", alpha = 0.8)

x_plot = np.linspace(-12.027530e-02,1.405552e-01).reshape(-1, 1)
y_plot = reg.predict(x_plot)

plt.plot(x_plot, y_plot,"r--", lw = 4)
plt.show()
```

```python
print("Entrenamiento", r2_score(y_train, y_train_predict))
print("Prueba", r2_score(y_test, y_test_predict))
```

Que la metrica r2 sea menor significa que el modelo se adapta de peor manera a los datos utilizados, una medida mayor de la r2 sugiere que el modelo representa de mejor manera los datos proporcionados.
