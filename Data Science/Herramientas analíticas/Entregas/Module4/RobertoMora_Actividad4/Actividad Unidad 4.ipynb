Para esta actividad es necesario un archivo con extensión csv llamado ___HR-Employee-Attrition-limpio.csv___ que debe estar en la misma carpeta en la que esta este archivo.

#### 1 - Lee el archivo __HR-Employee-Attrition.csv__ en un dataframe (1 pt).

```julia
using CSV
using DataFrames
df=CSV.read("HR-Employee-Attrition-limpio.csv",DataFrame)
first(df,5)
```

#### 2 - Importa StatsPlots , crea un histograma de la columna edad y guarda la grafica en una variable __p1__(2 pt).

```julia
using StatsPlots
theme(:default)
@df p1=df histogram(:Age,legend=false)
```

#### 3 - Crea un diagrama de cajas de la columna Age separado por la columna Gender y guarda la grafica en una variable __p2__ (2 pt).

```julia
@df p2=df boxplot(:Gender,:Age,legend=false)
```

#### 4 - Crea un diagrama de barras con el conteo total de personas por  la columna Age y guardala en __p3__ (2 pt).

```julia
p_group=groupby(df,:Age)
p_count=combine(p_group,nrow)
@df p3=p_count bar(:Age,:nrow,legend=false)
```

#### 5 - Aplica el tema :dark y vuelve a crear las gráficas (1 pt).

```julia
theme(:dark)
p1_dark = histogram(df[!, :Age], legend = false)
```

```julia
p2_dark = boxplot(df[!, :Gender], df[!, :Age], legend = false)
```

```julia
p_group = groupby(df, :Age)
p_count = combine(p_group, nrow)
p3_dark = bar(p_count[!, :Age], p_count[!, :nrow], legend = false)
```

#### 6 - Utiliza @layout para colocar la gráfica p1, p2 y p3 en una sola imagen (2 pt).

```julia
l = @layout [a b{0.5w} 
    c{0.5h}]
p = plot(p1_dark, p2_dark, p3_dark, layout = l, leg = false)
display(p)
```
