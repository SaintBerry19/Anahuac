# Temas

```julia
using Plots
gr();
```

```julia
data = Plots.fakedata(50, 4);
```

```julia
plot(data)
```

```julia
# https://github.com/JuliaPlots/PlotThemes.jl
themes = [:default :dark :ggplot2 :juno :lime :orange :sand :solarized :solarized_light :wong :wong2 :gruvbox_dark :gruvbox_light]
for theme in themes
    print(theme)
    display(Plots.showtheme(theme))
end
```

```julia
theme(:dark)
plot(data)
```

```julia
theme(:default)
plot(data)
```

# Colores

```julia
plot(data, w=3, palette=:inferno)
```

```julia
# http://juliagraphics.github.io/Colors.jl/stable/namedcolors/
plot(data[:,1], color=:coral2)
```

```julia
#RED-GREEN-BLUE
plot(data[:,1], color=RGB(0,1,0))
```

```julia
# https://htmlcolorcodes.com/color-picker/
plot(data[:,1], color=RGB(55/255,207/255,183/255))
```

```julia
plot(data[:,1], color=RGBA(1,0,0,0.5))
```

```julia
plot(data, color=[:red :blue :green :gray])
```

```julia
plot(data, color=[:red :blue :green :gray], background_color=RGB(0.24, 0.35, 0.33))
```

# Lineas

```julia
# Tipos de lineas
colors = [:red :blue :green :gray]
lines = [:scatter :steppre :steppost :sticks]
plot(data, color=colors, line=(3,lines), lab=map(string,lines))
# :path es la linea normal
```

```julia
# estilo de lineas
colors = [:red :blue :green :gray]
lines = [:solid :dash :dot :dashdot]
plot(data, color=colors, line=(3,lines), lab=map(string,lines))
```

```julia
markers = filter((m->begin
                m in Plots.supported_markers()
            end), Plots._shape_keys)
markers = reshape(markers, 1, length(markers))
n = length(markers)
x = (range(0, stop=10, length=n + 2))[2:end - 1]
y = repeat(reshape(reverse(x), 1, :), n, 1)
scatter(x, y, m=(8, :auto), lab=map(string, markers), xlim=(0, 10), ylim=(0, 10))
```

```julia
x = data[:,1]
y = data[:,2]
scatter(x,y, m = (6,:hexagon), color = :red)
```

```julia
plot(randn(100, 8), 
    layout=(4,2), 
    t=[:line :histogram :scatter :steppre :bar],
    bg_inside=[:orange :darkred :darkblue],
    bg = :grey,
    leg=false
)
```

```julia
l = @layout [  a{0.5w} [grid(2,2) 
                        b{0.3h} ]]

plot(randn(100, 6), 
    layout = l, 
    t = [:histogram :scatter :bar],
    bg_inside = [:orange :darkred :darkblue],
    bg = :grey,
    leg = false
)
```

```julia
p = plot(sin, [0], leg=false, xlims=[0,10π], ylims=[-1,1])
@gif for x = range(0, stop=10π, length=200)
    push!(p, x, sin(x))
end
```
