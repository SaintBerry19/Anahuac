```julia
using Plots
```

```julia
x = 1:10
y = rand(10)
```

```julia
plot(x,y)
savefig("plot.pdf")
```

```julia
p = plot(x,y, dpi=300, size(600,600))
savefig(p,"plot.png")
```

```julia
z = rand(10)
p = plot(x,y)
plot!(p,x,z)
```

```julia
y = rand(10,2)
p = plot(x,y, title="Nmero Aleatorio", labels=["azul" "naranja"], lw=3)
```

```julia
p = xlabel!("Eje x")
```

```julia
scatter(x,y,title = "scatter",label=["serie 1" "serie 2"])
```

```julia
display(p)
```

```julia
y = rand(10,8)
plot(x,y,layout=(2,2))
```

```julia
y = rand(10,1)
p1 = plot(x,y)
p2 = scatter(x,y)
p3 = bar(x,y)
p4 = histogram(x,y, fillcolor="red")
plot(p1,p2,p3,p4, layout=(2,2), legend=false)
```
