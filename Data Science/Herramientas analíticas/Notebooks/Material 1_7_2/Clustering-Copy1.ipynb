```julia
using CSV
using DataFrames
using Plots
using StatsPlots
```

```julia
df = DataFrame(CSV.File("IRIS.csv"))
first(df, 10)
```

```julia
@df df scatter(:petal_width, :petal_length ,legend=false)
```

```julia
using Clustering
```

```julia
df_reduced = df[:,[:petal_width,:petal_length]]
result = kmeans(transpose(Matrix(df_reduced)), 5) #Actualización
```

```julia
result.assignments
```

```julia
result.centers
```

```julia
result.totalcost
```

```julia
p = @df df scatter(
    :petal_width,
    :petal_length, 
    color = result.assignments,
    legend = false)
```

## Metodo del codo

```julia
errores = []
for i in 2:10
    result = kmeans(transpose(Matrix(df_reduced)), i)
    push!(errores,result.totalcost)
end
```

```julia
plot(errores,x = 2:10)
```

```julia
result = kmeans(transpose(Matrix(df_reduced)), 3)
@df df scatter(
    :petal_width,
    :petal_length, 
    color = result.assignments,
    legend = false)
```
